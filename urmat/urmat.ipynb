```python
## constants
M = 10000 # 20 # 20, 100, 1000
X = 1
h = 2*X/M
T:int = 5 # tau = 1 ?
gam = 1.4
# rho0 = 1
rho1 = 2
rho2 = 1
eps0 = 1
u0 = 0
```

```python
## formula functions
def get_p(rho, eps):
    return rho*eps*(gam-1)

def get_dp_drho(eps):
    return eps*(gam-1)

def get_dp_deps(rho):
    return rho*(gam-1)

def get_rho0(x):
    return x**2
    return (x>0)*rho1 + (x<=0)*rho2

def get_c(rho, eps):
    # print(rho, eps)
    return np.sqrt(gam*get_p(rho, eps)/rho)

def get_a(rho, u, eps):
    return np.array([
        [u, rho, 0],
        [1/rho * get_dp_drho(eps), u, 1/rho * get_dp_deps(rho)],
        [0, get_p(rho, eps)/rho, u]
    ])

def get_omega(rho, u, eps):
    rho_c = rho*get_c(rho, eps)
    p = get_p(rho, eps)
    dp_drho = get_dp_drho(eps)
    return np.array([
        [dp_drho, rho_c, get_dp_deps(rho)],
        [p, 0, -rho**2],
        [p, -rho_c, dp_drho]
    ])

def get_lambda(rho, u, eps):
    c = get_c(rho, eps)
    return np.diag((u+c, u, u-c))
```

```python
u = np.zeros((T, 2*M+1, 3)) # t, x, (rho, u, eps)
x = np.linspace(-X, X, 2*M+1) # M at each side + 0
u[0, :, 0] = get_rho0(x)
u[0, :, 1] = u0
u[0, :, 2] = eps0
```

```python
def calc():
    sigma = 1/h # tau/h
    for n in range(0, T-1):
        for m in range(0, 2*M+1):
            omega = get_omega(*u[n,m])
            inv_om = np.linalg.inv(omega)
            lamb = get_lambda(*u[n,m])
            if m != 0 and m != 2*M:
                u[n+1, m] = u[n,m] - sigma * (
                    0.5 * inv_om @ (lamb + np.abs(lamb)) @ omega @ (u[n,m-1] - u[n,m])
                    - 0.5 * inv_om @ (lamb - np.abs(lamb)) @ omega @ (u[n,m+1] - u[n,m])
                )
            elif m == 0:
                u[n+1, m] = u[n,m] - sigma * (
                    -inv_om @ (lamb - np.abs(lamb)) @ omega @ (u[n,m+1] - u[n,m])
                )
            else:
                u[n+1, m] = u[n,m] - sigma * (
                    0.5 * inv_om @ (lamb + np.abs(lamb)) @ omega @ (u[n,m-1] - u[n,m])
                )
```
